```python
import os
from dotenv import load_dotenv
load_dotenv()


groq_api_key=os.getenv("GROQ_API_KEY")
```

```python
from langchain_openai import ChatOpenAI
from langchain_groq import ChatGroq
model=ChatGroq( model="llama-3.3-70b-versatile",groq_api_key=groq_api_key)
model
```

```python
# English to Hindi translation using Groq API
from langchain_core.messages import HumanMessage,SystemMessage

messages=[
    SystemMessage(content="Translate the following from English to Hindi"),
    # gives message to model to translate the text from a Given language to Another language
    HumanMessage(content="Hello How are you?")
]

result=model.invoke(messages)
result
```

```python
from langchain_core.messages import HumanMessage,SystemMessage

messages=[
    SystemMessage(content="Translate the following from English to Bangali"),
    # gives message to model to translate the text from a Given language to Another language
    HumanMessage(content="India is a developing country with a rich cultural heritage and diverse population.")
]

result=model.invoke(messages)
result
```

```python
# English to Bangali translation using Groq API

# calling 'output_parsers' module from 'langchain_core' package to parse the output of the model into a string format 
# importing 'StrOutputParser' class from 'output_parsers' module
from langchain_core.output_parsers import StrOutputParser

# parses the output of the model into a string format
# parser - cuts down unnecessary information and returns only the required output
parser=StrOutputParser()
parser.invoke(result)
```

```python
# Chaining
chain = model|parser
chain.invoke(messages)
```

```python
# ARRANGING THE PROMPT FOR PROJECT 
### Prompt Templates
from langchain_core.prompts import ChatPromptTemplate

generic_template="Translate the following into {language}:"

prompt=ChatPromptTemplate.from_messages(
    [("system",generic_template),("user","{text}")]
    #system will take the text from user and convert it into the language specified by the user
)
```

```python
result=prompt.invoke({"language":"Bengali","text":"Hello"})
result
```

```python
# Chaining
chain = prompt|model|parser
chain.invoke({"language":"Bengali","text":"Hello How are you?"})
```
